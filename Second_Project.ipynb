###
Develop a Line chart using the functionality of pandas to show how automobile
sales fluctuate from year to year

```python
total_sales_per_year = df.groupby('Year')['Automobile_Sales'].sum()
print(total_sales_per_year)
```

```python

df_grouped = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

plt.figure(figsize=(20, 6))
plt.plot(df_grouped['Year'], df_grouped['Automobile_Sales'], marker='o', color='b', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales per Year')
plt.grid(True)
plt.show()
```

###
Develop a Bubble plot for displaying the impact of seasonality on

```python

groupby_Automobile_Sales_season=df.groupby('Seasonality_Weight')['Automobile_Sales'].sum().reset_index()
```

###
Create a pie chart to display the portion of advertising expenditure of
XYZAutomotives during recession and non-recession periods

```python
groupby_Advertising_Expenditure_Recession=df.groupby(['Advertising_Expenditure','Recession']).size().unstack().fillna(0)
groupby_Advertising_Expenditure_Recession.head()
```
