###
What is the maximum number of hours a person works per week (hours-per-
week feature)? How many people work such a number of hours and what is the
percentage of those who earn a lot among them?

```python
max_hours=df['hours-per-week'].max()
max_hours
```

```python
people_work_much=df[df['hours-per-week']>=99]
people_work_much
```

```python
max_salary=df['salary'].max()
max_salary
```

```python
num_high_earners= people_work_much[people_work_much['salary']==max_salary].shape[0]
num_high_earners
```

```python
total_people_work_much=people_work_much.shape[0]
total_people_work_much
```

```python
percentage=(num_high_earners/total_people_work_much)*100
percentage
```
